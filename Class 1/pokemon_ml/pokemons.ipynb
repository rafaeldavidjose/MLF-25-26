### Performance Basic Data Cleaning

```python
df.rename(columns={
        'Type 1': 'Type1', 
        'Type 2': 'Type2',
        'Sp. Atk': 'Sp_Atk',
        'Sp. Def': 'Sp_Def',
    }, inplace=True)

df
```

```python
na_counts = df.isna().sum()

na_counts
```

```python
df['Type2'].fillna('None', inplace=True)

df['Type2']
```

### Number of pokemons by generation

```python
df.groupby(['Generation'])['#'].count()
```

```python
df['Generation'].value_counts().sort_index()
```

### Number of pokemons by primary type

```python
df['Type1'].value_counts()
```

### Average stats by primary type

```python
df.groupby('Type1')[[
    'Total',
    'HP', 
    'Attack', 
    'Defense',
    'Sp_Atk', 
    'Sp_Def', 
    'Speed'
    ]].mean().sort_values("Total", ascending=False).round(1)
```

### Legendary vs Non-Legendary stats comparison

```python
df.groupby('Legendary')[[
    'Total',
    'HP', 
    'Attack', 
    'Defense',
    'Sp_Atk', 
    'Sp_Def', 
    'Speed'
    ]].mean().round(1)
```

### Histogram of the Primary Type Attack Stats

```python
sns.histplot(data=df, x='Attack', bins=20, kde=True)
```

### Histogram of the Speed attribute

```python
sns.histplot(data=df, x='Speed', bins=20, kde=True)
```

### Boxplot of Defense attribute by primary type (top 10 defenses)

Boxplot is a standardized way of displaying the distribution of data based on a five-number summary: minimum, 25th percentile (Q1), median (Q2), 75th percentile (Q3), and maximum. It can also indicate which observations, if any, might be considered outliers.

```python
top_types = df['Type1'].value_counts().head(10).index

top_types
```

```python
df[df['Type1'].isin(top_types)]
```

```python
defData = df[df['Type1'].isin(top_types)]

sns.boxplot(data=defData, x='Type1', y='Defense')
```
